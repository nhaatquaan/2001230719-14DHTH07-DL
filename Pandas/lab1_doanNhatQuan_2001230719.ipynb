```python
#11
# Tạo tensor x kích thước (3, 2) với giá trị nguyên trong khoảng [5, 10]
x = tf.random.uniform(shape=(3, 2), minval=5, maxval=11, dtype=tf.int32)

# Tạo tensor y kích thước (2, 3) với giá trị nguyên trong khoảng [15, 20]
y = tf.random.uniform(shape=(2, 3), minval=15, maxval=21, dtype=tf.int32)

print("Tensor x (3x2):\n", x.numpy())
print("Tensor y (2x3):\n", y.numpy())

# Lấy các giá trị trên đường chéo của x
diag_x = tf.linalg.diag_part(x)
print("\nGiá trị đường chéo của x:", diag_x.numpy())

# Thực hiện x + y
y_transposed = tf.transpose(y)
result_11 = x + y_transposed

print("\nKết quả x + transpose(y) (kích thước 3x2):\n", result_11.numpy())
```

```python
#12: Các hàm cơ bản trong Graph
@tf.function
def basic_ops(a, b):
    return a + b, a - b, a * b, a / b

#13, #14, #15, #16: Xây dựng Model tính toán
class GeometryModel(tf.Module):
    @tf.function
    def tinh_tong(self, x, y): #13
        return x + y

    @tf.function
    def tinh_hieu_tich(self, x, y): #14
        return x - y, x * y

    @tf.function
    def chu_vi_hcn(self, a, b): #15
        return (a + b) * 2

    @tf.function
    def dien_tich_hinh_hop(self, a, b, c): #16
        s_xung_quanh = (a + b) * c * 2
        s_toan_phan = (a * b) * 2 + s_xung_quanh
        return s_toan_phan

model = GeometryModel()
print("#15 Chu vi HCN (a=5, b=3):", model.chu_vi_hcn(tf.constant(5.0), tf.constant(3.0)).numpy())
```

```python
#19: Ví dụ mô hình Đường trung bình tam giác
@tf.function
def check_duong_trung_binh(MN, BC, AM, MB):
    # Điều kiện 1: M là trung điểm AB (AM = MB)
    is_midpoint = tf.equal(AM, MB)
    # Điều kiện 2: MN song song BC và MN = BC/2
    is_parallel_and_half = tf.logical_and(tf.equal(MN, BC/2), True) # Giả định song song là True

    # Kết luận NA = NC
    return tf.logical_and(is_midpoint, is_parallel_and_half)

# Chạy thử
conclusion = check_duong_trung_binh(tf.constant(5.0), tf.constant(10.0), tf.constant(4.0), tf.constant(4.0))
print("#19 Kết luận NA=NC là:", conclusion.numpy())
```
